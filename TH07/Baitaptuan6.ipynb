### W6A3

```python
num = list(map(int,input().split()))
k = int(input())
xoay = k
if k >= 0:
    while k != 0:
        for i in range(len(num)):
            if i <= xoay:
                num.append(num[0])
                num.remove(num[0])
                break
        k -= 1
else:
    while k != 0:
        for i in range(len(num)):
            if i >= xoay:
                num.insert(0,num[-1])
                num.pop(-1)
                break
        k += 1

print(tuple(num))
```

### W6A9

```python
dict1 = {'b': 20, 'a': 10, 'c': 30}
dict2 = {'a': 5, 'c': 15, 'd': 40}

res = {}

for key in dict1:
    res[key] = dict1[key]

for key in dict2:
    if key in res:
        res[key] = res[key] + dict2[key]
    else:
        res[key] = dict2[key]

all_keys = list(res.keys())

n = len(all_keys)
for i in range(n):
    for j in range(0, n - i - 1):
        if all_keys[j] > all_keys[j + 1]:
            all_keys[j], all_keys[j + 1] = all_keys[j + 1], all_keys[j]

sorted_res = {}
for key in all_keys:
    sorted_res[key] = res[key]

print(sorted_res)
```

### W6A10

```python
num = list(map(int,input().split()))
k = int(input())
res = []
n = len(num)
for i in range(n):
    for j in range(i+1,n):
        if num[i] + num[j] == k:
            res.append((num[i],num[j]))

res.sort(key = lambda x: x[0])

print(res)
```

### W6A12

```python
nums = [int(x) for x in input().split()]

if not nums:
    pass
else:
    counts = {}
    for n in nums:
        if n in counts:
            counts[n] += 1
        else:
            counts[n] = 1

    max_freq = 0
    for n in counts:
        if counts[n] > max_freq:
            max_freq = counts[n]

    res = None
    for n in counts:
        if counts[n] == max_freq:
            if res is None or n < res:
                res = n

    print(res)
```
