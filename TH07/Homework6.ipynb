```python
n = int(input("size n: "))

matrix = []
for i in range(n):
    row = list(map(int, input(f"Row {i}: ").split()))
    matrix.append(row)

chinh = []
for i in range(n):
    chinh.append(matrix[i][i])

phu = []
for i in range(n):
    phu.append(matrix[i][n - 1 - i])

print(*chinh) 
print(*phu)
```

### W6A18

```python
n, m, k = map(int, input("Nhập n, m, k: ").split())

matrix = []
for i in range(n):
    row = list(map(int, input(f"Nhập hàng {i}: ").split()))
    matrix.append(row)

if k < n:
    tong_hang_k = sum(matrix[k])
    print(f"Tổng các phần tử trên hàng {k}: {tong_hang_k}")
else:
    print(f"Không có hàng chỉ số {k}")

tong_cot_k = 0
if k < m:
    for i in range(n):
        tong_cot_k += matrix[i][k]
    print(f"Tổng các phần tử trên cột {k}: {tong_cot_k}")
else:
    print(f"Không có cột chỉ số {k}")
```

### W6A20

```python
def find_first_k(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

lst = list(map(int,input().split()))
k = int(input())
print(find_first_k(lst,k))
```

### W6A21

```python
def map_string_positions(s):
    words = s.split()
    d = {}
    for i, word in enumerate(words):
        if word not in d:
            d[word] = []
        d[word].append(i)
    return {word: tuple(pos) for word, pos in d.items()}

s = input().split()
print(map_string_positions(s))
```

### W6A22

```python
def find_max_index(lst):
    if not lst: return -1
    max_idx = 0
    for i in range(1, len(lst)):
        if lst[i] > lst[max_idx]:
            max_idx = i
    return max_idx

lst = list(map(int,input().split()))
print(find_max_index(lst))
```

### W6A23

```python
def delete_less_than_x(lst, x):
    return [item for item in lst if item >= x]

lst = list(map(int,input().split()))
x = int(input())
print(delete_less_than_x(lst,x))
```

### W6A24

```python
def move_zeros(lst, position='head'):
    zeros = [i for i in lst if i == 0]
    others = [i for i in lst if i != 0]
    return zeros + others if position == 'head' else others + zeros

lst = list(map(int,input().split()))
print(move_zeros(lst,position='head'))
```

### W6A25

```python
def missing_values(lst):
    full_set = set(range(10))
    current_set = set(lst)
    return list(full_set - current_set)

lst = list(map(int,input().split()))
print(missing_values(lst))
```

### W6A26

```python
def sum_local_maxima(lst):
    total = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            total += lst[i]
    return total

lst = list(map(int,input().split()))
print(sum_local_maxima(lst))
```

### W6A27

```python
def find_all_less_than_x(lst, x):
    indices = [i for i, val in enumerate(lst) if val < x]
    return indices if indices else -1

lst = list(map(int,input().split()))
x = int(input())
print(find_all_less_than_x(lst,x))
```

### W6A28

```python
def check_sequence(lst):
    is_increasing = True
    is_decreasing = True
    
    for i in range(len(lst) - 1):
        if lst[i] >= lst[i+1]: is_increasing = False
        if lst[i] <= lst[i+1]: is_decreasing = False
            
    if is_increasing: return "Thuc su tang"
    if is_decreasing: return "Thuc su giam"
    return "Vo danh"

lst = list(map(int,input().split()))
print(check_sequence(lst))
```
