### Linear Search
Thuật toán bắt đầu từ đầu danh sách và kiểm tra đối chiếu lần lượt từng phần tử một. Nếu khớp giá trị sẽ trả về vị trí (index), nếu duyệt hết mà không có sẽ trả về -1.

```python
def linear_search(lst,value):
    vitri = []
    for i in range(len(lst)):
        if lst[i] == value:
            vitri.append(i)
    if vitri == []:
        return -1
    return vitri

assert linear_search([30,11,70,40,41,70,57,52,25],70) == [2,5]
assert linear_search([30,11,70,40,41,70,57,52,25],99) == -1
print('DONE!')
```

### Binary Search
Tìm kiếm bằng cách liên tục chia đôi mảng dữ liệu để thu hẹp phạm vi

```python
def binary_search(lst,value):
    lst.sort()
    left,right = 0,len(lst)-1
    while left <= right:
        mid = (left+right)//2
        if lst[mid] == value:
            return mid
        elif lst[mid] < value:
            left = mid + 1
        else:
            right = mid - 1
    return -1

assert binary_search([14,11,40,30,25,41,57,52,70],57) == 7
assert binary_search([14,11,40,30,25,41,57,52,70],99) == -1
print('DONE!')
```

### Bubble Sort
Lặp lại việc so sánh và hoán đổi các cặp phần tử kề nhau nếu chúng bị ngược thứ tự.

```python
def bubble_sort(lst):
    for i in range(len(lst)):
        swapped = False
        for j in range(len(lst)-i-1):
            if lst[j] > lst[j+1]:
                lst[j],lst[j+1] = lst[j+1],lst[j]
                swapped = True
        if not swapped:
            break
    return lst


assert bubble_sort([30,11,70,45,41]) == [11,30,41,45,70]
print('DONE!')
```

### Selection Sort
Tìm kiếm phần tử nhỏ nhất trong phần chưa sắp xếp, sau đó hoán đổi nó với phần tử đầu tiên của phần này.

```python
def selection_sort(lst):
    for i in range(len(lst)):
        min_idx = i
        for j in range(i+1,len(lst)):
            if lst[j] < lst[min_idx]:
                min_idx = j
        lst[i],lst[min_idx] = lst[min_idx],lst[i]
    return lst
    
assert selection_sort([65,24,13,21,10]) == [10,13,21,24,65]
print('DONE!')
```

### Insertion Sort
Thuật toán sẽ duyệt và chèn lần lượt từng phần tử của danh sách chưa sắp xếp vào đúng vị trí tương ứng trong phần danh sách đã được sắp xếp.

```python
def insertion_sort(lst):
    for i in range(1,len(lst)):
        key = lst[i]
        j = i - 1
        while j >= 0 and lst[j] > key:
            lst[j+1] = lst[j]
            j -= 1
        lst[j+1] = key
    return lst

assert insertion_sort([12,11,14,7,8]) == [7,8,11,12,14]
print('DONE!')
```

### Merge Sort
Merge Sort phân chia mảng đầu vào thành hai nửa một cách đệ quy cho đến khi không thể chia nhỏ được nữa. Sau đó, các mảng con này sẽ được sắp xếp độc lập và "trộn" (merge) lại với nhau để tạo thành một mảng đã sắp xếp hoàn chỉnh.

```python
def merge_sort(lst):
    if len(lst) <= 1:
        return lst
    
    mid = len(lst) // 2
    left = lst[:mid]
    right = lst[mid:]
    
    sorted_left = merge_sort(left)
    sorted_right = merge_sort(right)
    
    return merge(sorted_left, sorted_right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
            
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result

assert merge_sort([39,25,40,7]) == [7,25,39,40]
print('DONE!')
```

### Quick Sort
Chọn một phần tử làm "trục" (pivot), sau đó phân chia các phần tử còn lại sang hai bên trục (nhỏ hơn sang trái, lớn hơn sang phải) rồi thực hiện gọi đệ quy.

```python
def partition(lst,low,high):
    pivot = lst[high]
    i = low - 1
    for j in range(low,high):
        if lst[j] <= pivot:
            i += 1
            lst[i],lst[j] = lst[j], lst[i]
    lst[i+1],lst[high]= lst[high],lst[i+1]
    return i + 1

def quick_sort(lst,low,high):
    if low < high:
        pi = partition(lst,low,high)
        quick_sort(lst,low,pi - 1)
        quick_sort(lst,pi + 1,high)
```
