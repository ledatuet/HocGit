### W5A16

```python
def Amstrong(so):
    so_saochep = so
    so_saochep1 = so
    n = 0
    tong = 0
    while so != 0: #xem n bang bao nhieu
        n += 1
        so //= 10
    
    while so_saochep != 0:
        a = so_saochep % 10
        tong += a**n
        so_saochep //= 10
    
    if tong == so_saochep1:
        return True
    return False

Amstrong(370)
```

### W5A18

```python
def palindrome(n):
    so_goc = n
    so_dao_nguoc = 0
    while n > 0:
        chu_so_cuoi = n % 10
        so_dao_nguoc = so_dao_nguoc * 10 + chu_so_cuoi
        n //= 10
    
    if so_goc == so_dao_nguoc:
        return True
    else:
        return False

palindrome(4554)
```

### W5A19

```python
def magical_num(n):
    while n >= 10:
        tong = 0
        while n > 0:
            chu_so_cuoi = n % 10
            tong += chu_so_cuoi
            n //= 10
        n = tong #quay lai tinh tiep
    if n == 1:
        return True
    else:
        return False

magical_num(19)
```

### W5A20

```python
def detect_number_type(s):
    temp_s = s
    if s.startswith('-'):
        temp_s = s[1:]
    if len(temp_s) == 0:
        return 'neither'

    is_int = True
    for char in temp_s:
        if not ('0' <= char <= '9'):
            is_int = False
            break
    
    if is_int:
        return 'int'

    dot_count = 0
    only_digits_and_dot = True
    
    for char in temp_s:
        if char == '.':
            dot_count += 1
        elif not ('0' <= char <= '9'):
            only_digits_and_dot = False
            break
            
    if dot_count == 1 and only_digits_and_dot:
        if temp_s == ".":
            return 'neither'
        return 'float'

    return 'neither'

print(detect_number_type("123"))    
print(detect_number_type("-456"))   
print(detect_number_type("3.14"))   
print(detect_number_type("abc"))    
print(detect_number_type("1.2.3")) 
```

### W5A25

```python
def tinh_dien_tich(a, b, c):
    s = 0.5 * abs(a[0]*(b[1] - c[1]) + b[0]*(c[1] - a[1]) + c[0]*(a[1] - b[1]))
    return s

def point_in_quadrilateral(quad, p):
    d0, d1, d2, d3 = quad
    
    s_lon1 = tinh_dien_tich(d0, d1, d2)
    s_nho1 = tinh_dien_tich(p, d0, d1) + tinh_dien_tich(p, d1, d2) + tinh_dien_tich(p, d2, d0)
    
    s_lon2 = tinh_dien_tich(d0, d2, d3)
    s_nho2 = tinh_dien_tich(p, d0, d2) + tinh_dien_tich(p, d2, d3) + tinh_dien_tich(p, d3, d0)

    if abs(s_lon1 - s_nho1) < 0.000001 or abs(s_lon2 - s_nho2) < 0.000001:
        return True
    else:
        return False

quad_vd = [(0,0), (4,0), (4,3), (0,3)]
p_vd = (2,1)

if point_in_quadrilateral(quad_vd, p_vd):
    print("Điểm nằm TRONG tứ giác")
else:
    print("Điểm nằm NGOÀI tứ giác")
```

### W5A28

```python
import math

x1, y1 = map(float, input("Nhập tọa độ A (x1 y1): ").split())
x2, y2 = map(float, input("Nhập tọa độ B (x2 y2): ").split())
x3, y3 = map(float, input("Nhập tọa độ C (x3 y3): ").split())

ux, uy = x1 - x2, y1 - y2  
vx, vy = x3 - x2, y3 - y2  

dot_product = ux * vx + uy * vy

mag_ba = math.sqrt(ux**2 + uy**2)
mag_bc = math.sqrt(vx**2 + vy**2)


if mag_ba == 0 or mag_bc == 0:
    print("Không tạo thành góc vì có điểm trùng nhau!")
else:
    cos_angle = dot_product / (mag_ba * mag_bc)

    cos_angle = max(-1.0, min(1.0, cos_angle))

    angle_radian = math.acos(cos_angle)
    angle_degree = math.degrees(angle_radian)

    print(f"Góc ABC là: {angle_degree:.2f} độ")
```

```python
def tim_max_min(a, b, c):
    
    max_val = a
    min_val = a
    
    if b > max_val:
        max_val = b
    if b < min_val:
        min_val = b
        
    if c > max_val:
        max_val = c
    if c < min_val:
        min_val = c
        
    return [max_val, min_val]

ket_qua = tim_max_min(10, 5, 20)
print(ket_qua)
```

### W5A30

```python
def nhithuc(n,k):    
    #tinh n!
    gt_n = 1
    for i in range(1, n + 1):
        gt_n = gt_n * i
    
    #tinh k!
    gt_k = 1
    for i in range(1, k + 1):
        gt_k = gt_k * i
        
    #tinh (n-k)!
    gt_n_k = 1
    for i in range(1, (n - k) + 1):
        gt_n_k = gt_n_k * i
    
    #c(n,k) = n! / k! * (n-k)!
    c_n_k = gt_n // (gt_k * gt_n_k)

    return c_n_k

nhithuc(10,5)
```

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):

    if km <= 1:
        tien_di_chuyen = 12000
    else:
        tien_di_chuyen = 12000 + (km - 1) * 15000
    
    tien_cho = wait_minutes * 500
    
    tong_tien = tien_di_chuyen + tien_cho
    
    if start_hour >= 22 or start_hour <= 5:
        tong_tien = tong_tien * 1.1  
        
    if km > 30:
        tong_tien = tong_tien * 0.95 
        
    return tong_tien

ket_qua = smart_taxi_fare(4, 23, 35, 10)
print(ket_qua)
```

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    
    tong = 0
    for i in range(1, n + 1):
        s = str(i) 
        co_lap_lai = False
        
        for chu_so in s:
            if s.count(chu_so) > 1:
                co_lap_lai = True
                break
        
        if co_lap_lai == False:
            tong += i
    return tong


sum_1_to_n(25)
```
