### R1E1

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    fare_move = 0
    
    if car_type == 4:
        if km <= 0.8:
            fare_move = 11000
        elif km <= 20:
            fare_move = 11000 + (km - 0.8) * 13000
        else:
            fare_move = 11000 + (19.2 * 13000) + (km - 20) * 10500
            
    elif car_type == 7:
        if km <= 0.8:
            fare_move = 12000
        elif km <= 20:
            fare_move = 12000 + (km - 0.8) * 14000
        else:
            fare_move = 12000 + (19.2 * 14000) + (km - 20) * 12000

    fare_wait = 0
    if wait_minutes > 5:
        extra_minutes = wait_minutes - 5
        blocks = (extra_minutes + 4) // 5 
        fare_wait = blocks * 2000

    subtotal = fare_move + fare_wait
    is_night = False

    if start_hour >= 22 or start_hour <= 5:
        subtotal = subtotal * 1.2
        is_night = True

    total = subtotal
    
    if km > 30 and not is_night:
        total = total * 0.9
        
    if total < 50000 and km > 5:
        total = 50000
        
    return int(total)

car_type, start_hour, km, wait_minutes = map(float,input().split())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

### R1E14

```python
def nhithuc(n,k):
    if k < 0 or k > n:
        return 0
    
    if k == 0 or k == n:
        return 1
    
    gtn = 1
    for i in range(1,n+1):
        gtn *= i

    gtk = 1
    for j in range(1,k+1):
        gtk *= j
    
    gtnk = 1
    for so in range(1,(n-k)+1):
        gtnk *= so
    
    c_n_k = int(gtn/(gtk*gtnk))
    return c_n_k

n,k = map(int,input().split())
print(nhithuc(n,k))
```

```python
def returnmaxmin(a,b,c):
    x = [a,b,c]
    max = x[0]
    min = x[0]
    for i in x:
        if i > max:
            max = i
    
    for j in x:
        if j < min:
            min = j
    
    return [max,min]

a,b,c = map(float,input().split())
print(returnmaxmin(a,b,c))
```

### R1E25

```python
def point_in_quadrilateral(quad, p):
    x, y = p
    inside = False
    n = len(quad)

    for i in range(n):
        p1_x, p1_y = quad[i]
        p2_x, p2_y = quad[(i + 1) % n]
        
        if min(p1_y, p2_y) < y <= max(p1_y, p2_y):
            if x <= max(p1_x, p2_x):
                if p1_y != p2_y:
                    x_intersect = (y - p1_y) * (p2_x - p1_x) / (p2_y - p1_y) + p1_x
                
                if p1_x == p2_x or x <= x_intersect:
                    inside = not inside
                    
    return inside
```

### R1E30

```python
def detect_number_type(s):
    temp_s = s
    if temp_s.startswith('+') or temp_s.startswith('-'):
        temp_s = temp_s[1:]
    
    if len(temp_s) == 0:
        return 'neither'

    dot_count = temp_s.count('.')
    
    if dot_count == 0:
        if temp_s.isdigit():
            return 'int'
        else:
            return 'neither'
            
    elif dot_count == 1:
        parts = temp_s.split('.')
        left = parts[0]
        right = parts[1]
        
        if left.isdigit() and right.isdigit():
            return 'float'
        else:
            return 'neither'
    
    else:
        return 'neither'

s = input()
print(detect_number_type(s))
```
