### W5A3

```python
def isprime(n):
    SNT = True
    if n > 3:
        for i in range(2,int(n**0.5)+1):
            if n % i == 0:
                SNT = False
                break
        if SNT:
            print('La SNT')
        else:
            print('Khong la SNT')
    elif n == 2 or n == 3:
        print('La SNT')
    else:
        print('Khong la SNT')

isprime(7)
```

```python
def Perfect_Num(n): #tong uoc = so do
    s = 0
    for i in range(1,n):
        if n % i == 0: #kiemtrauoc
            s += i
    if s == n:
        print(n)

Perfect_Num(8)
```

### W5A7

```python
def calcu(a,toantu,b):
    if toantu == '+':
        x = float(a + b)
        print(f'Tong: {a} + {b} = {x:.2f}')
    
    if toantu == '-':
        x = float(a-b)
        print(f'Hieu: {a} - {b} = {x:.2f}')
    
    if toantu == '*':
        x = float(a*b)
        print(f'Tich: {a} * {b} = {x:.2f}')
    
    if toantu == '/':
        x = float(a/b)
        print(f'Thuong: {a} / {b} = {x:.2f}')

calcu(1,'-',2)
```

### W5A8

```python
def hamming(x,y):
    dif = x ^ y
    kc = 0
    while dif > 0:
        if dif % 2 == 1:
            kc += 1
        dif //= 2
    return kc
a = 1
b = 4
result = hamming(a,b)
print(result)
```

### W5A9

```python
def nhap(n):
    s = 0
    while n != 0:
        a = n % 10
        s += a
        n //= 10
    
    print(s)

nhap(123)
```

### W5A10

```python
def check_isomorphic(s, t):
    mapping_s_to_t = {}
    used_in_t = {}

    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]

        if char_s in mapping_s_to_t:
            if mapping_s_to_t[char_s] != char_t:
                return False
        else:
            if char_t in used_in_t:
                return False
            mapping_s_to_t[char_s] = char_t
            used_in_t[char_t] = True

    return True


print(check_isomorphic("abca", "zbxz")) 
print(check_isomorphic("foo", "bar")) 
```

### W5A12

```python
def divide_with_remainder(dividend, divisor):
    q = dividend // divisor

    r = dividend % divisor
    
    return q,r

result = divide_with_remainder(13, 3)
print(result) 
```

```python
def compute(a,b,op):
    if op == '+':
        print(f'op la +: {a} + {b} = {a+b}')

    if op == '-':
        print(f'op la -: {a} - {b} = {a-b}')

    if op == '*':
        print(f'op la *: {a} * {b} = {a*b}')

    if op == '/':
        print(f'op la /: {a} / {b} = {a/b}')
    
compute(1,1,'+')
```
