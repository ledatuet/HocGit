### W4A25

```python
max = None
min = None
while True:
    n = int(input())
    if n == -1:
        break
    if max == None:
        max = n
        min = n
    else:
        if max < n:
            max = n
        if min > n:
            min = n

if max != None:
    print(max)
    print(min) 
```

### W4A26

```python
A = int(input())
f1 = 1
f2 = 1
if A < 1:
    print('Not any suitable Fibo')
else:
    f_next = f1 + f2
    while f_next <= A:
        f1 = f2
        f2 = f_next
        f_next = f1 + f2
    print(f2)
```

### W4A27

```python
s = input()
length = 0
for char in s:
    length += 1

word_count = 0
in_word = False 

for i in range(length):
    char = s[i]
    
    if char != ' ': 
        if in_word == False:
            word_count += 1
            in_word = True
    else:
        in_word = False

print(word_count)
```

```python
nhap = input()
hoa = 0
thuong = 0
so = 0
for i in nhap:
    if 'A' <= i <= 'Z':
        hoa += 1
    if 'a' <= i <= 'z':
        thuong += 1
    if '0' <= i <= '9':
        so += 1
print(f'Hoa: {hoa}')
print(f'Thuong: {thuong}')
print(f'So: {so}')
```
