### W4A4 COUNTINGDIGITS

```python
n = int(input())
dem = 0
if n == 0:
    dem = 1
    print(dem)
elif n > 0:
    while n > 0:
        n = n // 10
        dem += 1
    print(dem)
else:
    n = (-1)*n
    while n > 0:
        n = n // 10
        dem += 1
    print(dem)
```

### W4A8 REVERSEADDPALINDROME

```python
def nhap(n):
    a = str(n)
    l = len(a)
    for i in range(0,l//2):
        if a[i] != a[l-1-i]:
            return False
        return True
def nguoc(n):
    k = 0
    while n > 0:
        k = k*10 + n%10
        n //= 10
    return k
n = int(input())
k = 0
while n < 10000:
    if nhap(n):
        print(k,n)
    n = n+nguoc(n)
    k += 1
```

### W4A10 LONGESTCOLLATZUPTON

```python
def nhap(n):
    c = 1
    while n > 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3*n + 1
        c += 1
    return c
n = int(input())
k = 0
x = 0
for i in range(1,n+1):
    if nhap(i) > k:
        k = nhap(i)
        x = i
print(x,k)
```

### W4A11 EVENDIVISORS

```python
def nhap(n):
    if n % 2 != 0:
        return 0
    c = 0
    for i in range(1,int(n**0.5)+1):
        if n % i == 0:
            if i % 2 == 0:
                c += 1
            if (n // i) % 2 == 0 and i != (n // i):
                c += 1
    return c
n = int(input())
print(nhap(n))
```

### W4A13 CLOSEPAIRNUMBER

```python
def nhap(n):
    c = 0
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            c += i
            if (n // i) != i:
                c += n//i
    return c+1
a,b = map(int,input().split())
if nhap(a) == b and nhap(b) == a:
    print('True')
else:
    print('False')
```

### W4A14 GCD

```python
m,n = list(map(int, input().split()))
k = 1
while m % 2 == 0 and n % 2 == 0:
    k *= 2
    m //= 2
    n //= 2
a = min(m,n)
for i in range(3,a+1,2):
    while m % i == 0 and n % i == 0:
        k *= i
        m //= i
        n //= i
print(k)
```
