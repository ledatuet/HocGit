```python
a, b, c = map(float,input().split())
if (a+b>c) and (a+c>b) and (b+c>a):
    if a == b == c:
        print('Deu')
    elif (a == b) or (a == c) or (b == c):
        print('Can')
    else:
        if (((a**2) + (b**2))**0.5) == c or (((a**2) + (c**2))**0.5) == b or (((c**2) + (b**2))**0.5) == a:
            print('Vuong')
        else:
            print('Thuong')

    cv = a + b + c
    p = (a+b+c)/2
    dt = (p * (p - a) * (p - b) * (p - c))**0.5

    print(f'Chu vi: {cv}')
    print(f'Dien tich: {dt}')
    
else:
    print('Khong la ba canh cua mot tam giac')
```

### Bai5

```python
a,b,c = map(int,input().split())
if a == 0: #bx+c=0
    if b == 0 and c != 0:#c=0
        print('Vo nghiem')
    elif b == 0 and c == 0:#0=0
        print('Vo so nghiem')
    else:#bx+c=0
        x = float(-c/b)
        print(f'Nghiem duy nhat: {x}')
else:#ax^2+bx+c=0
    delta = (b**2) - (4*a*c)
    if delta < 0:
        print('Vo nghiem')
    elif delta == 0:
        x = float((-b)/(2*a))
        print(f'Nghiem kep: x1 = x2 = {x}')
    else: #delta > 0
        x1 = float((-b + delta**0.5)/(2*a))
        x2 = float((-b - delta**0.5)/(2*a))
        print('2 nghiem phan biet')
        print(f'x1 = {x1}')
        print(f'x2 = {x2}')
```
