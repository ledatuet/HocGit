### W3A16

```python
n = float(input())
if n == int(n):
    tronlen = int(n)
    tronxuong = int(n)
    gannhat = int(n)
else:
    if n > 0:
        tronlen = int(n) + 1
        tronxuong = int(n)
        if n - int(n) < 0.5:
            gannhat = int(n)
        else:
            gannhat = int(n) + 1
    else:
        tronlen = int(n)
        tronxuong = int(n) - 1
        if n - (int(n) - 1) > 0.5:
            gannhat = int(n)
        else:
            gannhat = int(n) - 1
    
print(f'{tronlen} {tronxuong} {gannhat}')
```
